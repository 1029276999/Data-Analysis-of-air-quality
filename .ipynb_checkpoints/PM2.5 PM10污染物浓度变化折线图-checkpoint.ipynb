```python
import csv
from pylab import mpl
import matplotlib.pyplot as plt
from datetime import datetime
mpl.rcParams['font.sans-serif'] = ['SimHei'] #SimHei
 
filename=r'nanchang.csv'
 
#csv.reader(),fnext
#csvnext(),f
#
with open(filename) as f:
    reader=csv.reader(f)
    header_row=next(reader)
    dates,PM2s,PM10s,So2s, No2s, Cos, O3s=[],[],[],[],[],[],[]
    i=1
    for row in reader:
        i=i+1
        if i%2==1and i>3048:
            date=datetime.strptime(row[0],'%Y-%m-%d')
            PM2=int(row[4])
            PM10=int(row[5])
            So2= int(row[6])
            No2= int(row[7])
         #   Co=int(row[8])
            O3=int(row[9])

            dates.append(date)
            PM2s.append(PM2)
            PM10s.append(PM10)
            So2s.append(So2)
            No2s.append(No2)
          #  Cos.append(Co)
            O3s.append(O3)
         
#print(dates)       
fig=plt.figure(dpi=128,figsize=(15,6))
plt.plot(dates,PM2s,c='red',alpha=0.5,label="PM2.5")
plt.plot(dates,PM10s,c='blue',alpha=0.5,label="PM10")
#plt.plot(dates,So2s,c='yellow',alpha=0.5,label="")
#plt.plot(dates,No2s,c='pink',alpha=0.5,label="",)
#plt.plot(dates,Cos,c='black',alpha=0.5,label="")
#plt.plot(dates,O3s,c='orange',alpha=0.5,label="")
plt.legend(loc=2)
#
#plt.fill_between(dates,AQIindexs,AQIranks,facecolor='blue',alpha=0.1)
 
plt.title('PM2.5PM10-2018',fontsize=24)
plt.xlabel('',fontsize=12)
plt.grid(True)
#
fig.autofmt_xdate()
plt.ylabel('',fontsize=10,fontproperties='SimHei')
plt.tick_params(axis='both',labelsize=12)
 
#
#x_ticks=dates[::3]
#plt.xticks(x_ticks)
#y_ticks=list(range(45,75,5))
#plt.yticks(y_ticks)
 
plt.show()
```
